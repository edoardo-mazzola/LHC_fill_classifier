```python
import pickle
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

INTERIM_DATA_PATH = '../data/interim/'
interim_file_path = os.path.join(INTERIM_DATA_PATH, 'all_fills_time_series.pkl')

print(f"Caricamento del file di serie temporali pulite da: {interim_file_path}")

try:
    with open(interim_file_path, 'rb') as f:
        # Questo caricherà il dizionario {'Fill_ID': {'Relative_Time': array, 'Luminosity': array}}
        cleaned_fills_time_series = pickle.load(f)
    print(f"Caricamento completato. Numero di fill caricati: {len(cleaned_fills_time_series)}")
except FileNotFoundError:
    print(f"Errore: Il file '{interim_file_path}' non è stato trovato.")
    print("Assicurati di aver eseguito lo script in utils.py per generarlo.")
    cleaned_fills_time_series = {} # Inizializza un dizionario vuoto per evitare errori successivi
except Exception as e:
    print(f"Si è verificato un errore durante il caricamento del file: {e}")
    cleaned_fills_time_series = {}
print("Fill Numbers:", cleaned_fills_time_series.keys())
```

```python
Year=2022
Fill_Number = 8124

fill_data = cleaned_fills_time_series[Fill_Number]
luminosity_values = fill_data['Luminosity']
time_values = fill_data['Relative_Time']


plt.figure(figsize=(10,6))

# Cleaned luminosity
plt.plot(time_values, luminosity_values,
         label='Cleaned Luminosity',
         linestyle='-', color='orange')



plt.xlabel('Time [H]')
plt.ylabel(r'Luminosity [nb$^{-1}$]')
plt.title(f'Fill {Fill_Number} - Year {Year}')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#derivazione dei tlev dai set di dati
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Liscio i dati
L_smooth = savgol_filter(luminosity_values, window_length=11, polyorder=3)

# Calcolo derivata smussata
dLdt = np.gradient(L_smooth, time_values)

# Definisco una soglia per dire che "qui inizia a calare"
soglia = 1e3  
window = 200    # numero di punti consecutivi con derivata negativa

for i in range(len(dLdt) - window):
    if np.all(dLdt[i:i+window] < soglia):
        t0 = time_values[i]
        print(f"Inizio decadimento trovato a t = {t0:.2f}")
        break
else:
    print("Nessun punto di inizio decadimento trovato.")

# Visualizza
plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(time_values, luminosity_values, label='Dati originali')
plt.plot(time_values, L_smooth, label='Smussato')
plt.axvline(t0, color='red', linestyle='--', label=f'Inizio decadimento: t={t0:.2f}')
plt.legend()
plt.title("Luminosità vs Tempo")

plt.subplot(2,1,2)
plt.plot(time_values, dLdt, label="Derivata")
plt.axhline(soglia, color='gray', linestyle=':')
plt.axvline(t0, color='red', linestyle='--')
plt.legend()
plt.title("Derivata della luminosità")
plt.xlabel("Tempo")

plt.tight_layout()
plt.show()
```

```python
# Parametri per "inizio plateau"
import numpy as np

print("Range tipici di dLdt:")
print(f"  max: {np.max(dLdt)}")
print(f"  min: {np.min(dLdt)}")
print(f"  std: {np.std(dLdt)}")

# Parametri regolabili
flat_mean_threshold = 1e4  # massimo valore medio assoluto accettabile
flat_std_threshold = 1e4  # massimo rumore accettabile
flat_window = 100          # numero di punti per considerare un tratto stabile

t_flat = None

for i in range(len(dLdt) - flat_window):
    window = dLdt[i:i+flat_window]
    mean_abs = np.mean(np.abs(window))
    std_dev = np.std(window)

    if mean_abs < flat_mean_threshold and std_dev < flat_std_threshold:
        t_flat = time_values[i]
        print(f"Inizio plateau (piatto stabile) trovato a t = {t_flat:.2f}")
        break
else:
    print("Nessun plateau stabile trovato con i parametri attuali.")


# Visualizzazione aggiornata
plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(time_values, luminosity_values, label='Dati originali')
plt.plot(time_values, L_smooth, label='Smussato')
if t0:
    plt.axvline(t0, color='red', linestyle='--', label=f'Inizio decadimento: t={t0:.2f}')
if t_flat:
    plt.axvline(t_flat, color='green', linestyle='--', label=f'Inizio plateau: t={t_flat:.2f}')
plt.legend()
plt.title("Luminosità vs Tempo")

plt.subplot(2,1,2)
plt.plot(time_values, dLdt, label="Derivata")
plt.axhline(soglia, color='gray', linestyle=':', label='Soglia decadimento')
plt.axhline(flat_mean_threshold, color='green', linestyle=':', label='Soglia plateau')
plt.axhline(-flat_mean_threshold, color='green', linestyle=':')
if t0:
    plt.axvline(t0, color='red', linestyle='--')
if t_flat:
    plt.axvline(t_flat, color='green', linestyle='--')
plt.legend()
plt.title("Derivata della luminosità")
plt.xlabel("Tempo")

plt.tight_layout()
plt.show()
```
